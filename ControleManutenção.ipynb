# Controle de manutenção

Programa para ler e identificar possíveis pontos de superaquecimento


### Importar a base de dados


```python
import pandas as pd 

controle = pd.read_excel('ControleTemperaturaNovo.xlsx')
display(controle)
```

```python
controle_ordenado = controle.sort_values(by='Temp')
display(controle_ordenado)
```

### Filtrar para apenas os pontos quentes

```python
PontosQuentes = controle_ordenado.loc[controle_ordenado['Temp']>89 ]
display(PontosQuentes)
```

### Criar uma lista com os pontos quentes

```python
lista_pontos = PontosQuentes['Ponto'].tolist()
display(lista_pontos)
```

### Envio automático de email com os pontos quentes

```python
import win32com.client as win32

outlook = win32.Dispatch('outlook.application')

email = outlook.CreateItem(0)

email.To = "velocitypincho@gmail.com"
email.Subject = "ALERTA!"
email.HTMLBody = f"""
<p>Olá controle </p>
<p>Percebeu-se que os pontos {lista_pontos} estão sobreaquecidos e correm risco de gerar um curto-circuito!</p>

<p>Recomendo enviar uma equipe de manutenção com urgência </p>
<p>Atenciosamente </p>
<p>Código Python </p>
"""
email.Send()
print('email enviado')
```
